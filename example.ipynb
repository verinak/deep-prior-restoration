# Denoising

```python
from PIL import Image
from IPython.display import display

from dip_denoising import DenoisingModel

img_path = './imgs/F16_noisy.png'
image = Image.open(img_path) # pil image
display(image)
```

```python
# optimize
model = DenoisingModel(print_iter=100) # set print_iter=None to hide output
output_pil = model.denoise(img_pil=image)

# show output image
display(output_pil)
```

# Inpainting

```python
from PIL import Image
from IPython.display import display

from dip_inpainting import InpaintingModel

img_path = './data/img_resized.png'
mask_path = './data/img_mask.png'

image = Image.open(img_path)
mask = Image.open(mask_path)

display(image)
display(mask)
```

```python
# optimize
model = InpaintingModel(print_iter=100)
output_pil = model.inpaint(img_pil=image, mask_pil=mask)

# show output image
display(output_pil)
```

# Super Resolution

```python
from PIL import Image
from IPython.display import display

from dip_superresolution import SuperResolutionModel

img_path = './data/cat.jpg'
image = Image.open(img_path)

display(image)
```

```python
# optimize
model = SuperResolutionModel(print_iter=100)
output_pil = model.upscale(img_pil=image)

# show output image
display(output_pil)
```
